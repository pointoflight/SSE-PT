```python
import random
```

```python
file1 = open("ml1m.txt","r")
lines = file1.readlines()
lines
```

```python
len(lines)
```

```python
file2 = open("ml1m-s.txt","w")
l = []

i = 0

while i < len(lines):
    print(i)
    li = []
    cur = lines[i].split(' ')
    cur_u = cur[0]
    
    j = i + 1
    
    while j < len(lines):
        nex = lines[j].split(' ')
        if nex[0] != cur_u:
            break
        j += 1
        
    for k in range(i,j):
        li.append(lines[k].split(' ')[1])
    i = j
    
    l.append(li)
    
# l
```

```python
len(l)
```

```python
for i in range(len(l)):
    random.shuffle(l[i])
```

```python
lol = []

for i in range(len(l)):
    for j in range(len(l[i])):
        lol.append(str(i+1) + ' ' + l[i][j])

lol
```

```python
len(lol)
```

```python
file2.writelines(lol)
file2.close()
```
